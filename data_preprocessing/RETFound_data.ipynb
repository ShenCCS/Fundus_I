# Move file togeter

```python
import os
import shutil


source_folder = '/home/john/network/RETFound/data/1201_NewData/'
destination_folder = '/home/john/network/RETFound/data/1201_data'


def move_file(source_folder,destination_folder):
    count = 0
    for root,sub,files in os.walk(source_folder):
        for file in files:
            source_folder_path = os.path.join(root,file)
            destination_file_path = os.path.join(destination_folder,file)
            shutil.copy2(source_folder_path,destination_file_path)
            count += 1
    print("Image files number: ",count)

move_file(source_folder,destination_folder)
```
